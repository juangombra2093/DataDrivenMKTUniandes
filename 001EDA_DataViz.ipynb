# **Analisis Grafico** ðŸ“Š

```python
#Boxplot
plt.figure(figsize=(8, 6))  # Ajusta el tamaÃ±o del grÃ¡fico si es necesario
sns.boxplot(data=df, x='Income')
plt.title('Boxplot de Income')
plt.xlabel('Income')
plt.show()
```

```python
#Diagrama de densidad
plt.figure(figsize=(8, 6))  # Ajusta el tamaÃ±o del grÃ¡fico si es necesario
sns.kdeplot(data=df, x='Income', fill=True, color='blue', common_norm=False)
sns.histplot(data=df, x='Income', kde=True, color='blue', bins=30)
plt.title('Diagrama de Densidad de Income')
plt.xlabel('Income')
plt.show()
```

```python
# Crear el grÃ¡fico de barras
df['Kidhome'].value_counts().plot(kind='bar')

# Agregar tÃ­tulo y etiquetas
plt.title('DistribuciÃ³n de Kidhome')
plt.xlabel('NÃºmero de niÃ±os en el hogar')
plt.ylabel('Frecuencia')

# Mostrar el grÃ¡fico
plt.show()
```

```python
# Crear el diagrama de dispersiÃ³n
plt.figure(figsize=(8, 6))  # Ajusta el tamaÃ±o del grÃ¡fico si es necesario
sns.scatterplot(data=df, x='MntTotal', y='Income', color='blue', alpha=0.5)  # Diagrama de dispersiÃ³n
sns.regplot(data=df, x='MntTotal', y='Income', color='red', scatter=False)  # LÃ­nea de tendencia
plt.title('Diagrama de DispersiÃ³n entre MntTotal e Income con LÃ­nea de Tendencia')
plt.xlabel('MntTotal')
plt.ylabel('Income')
plt.show()
```

# **Matriz de Correlacion**

```python
# Lista con los nombres de las columnas numÃ©ricas que te interesan
columnas_a_incluir = ['Income','MntMeatProducts','MntWines','MntFruits',
                      'MntFishProducts','MntSweetProducts','MntGoldProds']

# Filtrar el DataFrame utilizando la lista
df_numeric = df[columnas_a_incluir]
```

```python
corr_matrix = df_numeric.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de CorrelaciÃ³n')
plt.show()
```
